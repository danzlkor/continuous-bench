```python
def remove_entries(input_array, index):
    # Create a boolean mask to select entries to keep
    array = np.copy(input_array)
    mask = np.ones(array.shape, dtype=bool)
    mask[:,index] = False
    
    # Filter the array based on the mask
    filtered_array = array[mask]
    
    return filtered_array
```

```python
import numpy as np

def similarity_matrix(y, x):
    # Compute the norms of y and x along the last dimension
    y_norm = np.linalg.norm(y, axis=2, keepdims=True)
    x_norm = np.linalg.norm(x, axis=2, keepdims=True)
    
    # Compute the dot product between y and x for all voxels and parameters/phenotypes simultaneously
    dot_product = np.matmul(y, np.transpose(x, (0, 2, 1)))
    
    # Compute the cosine similarity between y and each x vector for all voxels and parameters/phenotypes simultaneously
    cosine_sim = dot_product / np.matmul(y_norm, np.transpose(x_norm, (0, 2, 1)))
    
    # Return the cosine similarity matrix
    return cosine_sim
```

```python
y_array = np.random.rand(1000, 11, 5)
x_array = np.random.rand(1000, 100, 5)

sim_matrix = similarity_matrix(y_array, x_array)
```
